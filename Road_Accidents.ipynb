# Violation Data

```python
violation=pd.read_csv("/Users/apoorvi_sahu/Downloads/Road Accidents Violation.csv")
```

```python
violation.info()
```

```python
violation.rename(columns={"Category":"Cause"},inplace=True)
```

```python
violation=violation.iloc[:-1,:-3]
violation=violation.iloc[::2]
```

```python
violation=violation.reset_index(drop=True)
violation
```

```python
violation.isnull().sum()
```

```python
violation_long=violation.melt(id_vars="Cause",var_name="Year_Metric",value_name="Value")
violation_long
```

```python
violation_long[["Year","Metric"]]=violation_long["Year_Metric"].str.split("-",expand=True)
```

```python
violation_long
```

```python
violation_long["Year"]=violation_long["Year"].astype(int)
violation_long["Value"]=pd.to_numeric(violation_long["Value"].str.replace(",","",regex=True), errors="coerce")
```

```python
violation_long
```

```python
violation_final=violation_long.pivot_table(index=["Cause", "Year"],columns="Metric",values="Value",aggfunc="sum").reset_index()
violation_final
```

```python
violation_final.columns.name = None
violation_final=violation_final.rename(columns={"Killed":"Fatalities","injured":"Injuries"})
violation_final
```

```python
violation_final.to_csv("cleaned_violation.csv",index=False)
```

# Collision Data

```python
collision=pd.read_csv("/Users/apoorvi_sahu/Downloads/Road Accidents Collision.csv")
```

```python
collision
```

```python
collision=collision.iloc[:-1,1:-3]
collision=collision.iloc[::2]
```

```python
collision=collision.reset_index(drop=True)
collision
```

```python
collision.isnull().sum()
```

```python
collision=collision.rename(columns={"Type of collision":"Collision_Type"})
```

```python
collision_long=collision.melt(id_vars="Collision_Type",var_name="Year_Metric",value_name="Value")
collision_long
```

```python
collision_long
```

```python
collision_long["Year"]=collision_long["Year"].astype(int)
```

```python
collision_final=collision_long.pivot_table(index=["Collision_Type","Year"],columns="Metric",values="Value",aggfunc="sum").reset_index()
collision_final
```
